6. Для каждой страны отобразим общую сумму привлечённых инвестиций, которые получили компании, зарегистрированные в этой стране. Страну, в которой зарегистрирована компания, можно определить по коду страны. Отсортируем данные по убыванию суммы.

```python
query = """
SELECT country_code, 
       SUM(CAST(funding_total as float)) as total
FROM company
GROUP BY country_code
HAVING SUM(CAST(funding_total as float)) > 0
ORDER BY SUM(CAST(funding_total as float)) DESC
"""
pd.read_sql(query, con = engine).head(5)
```

7. Составим таблицу, в которую войдёт дата проведения раунда, а также минимальное и максимальное значения суммы инвестиций, привлечённых в эту дату. Оставим в итоговой таблице только те записи, в которых минимальное значение суммы инвестиций не равно нулю и не равно максимальному значению.

```python
query = """
SELECT CAST(funded_at AS date),
       MAX(raised_amount),
       MIN(raised_amount)
FROM funding_round
GROUP BY funded_at
HAVING MIN(raised_amount) != 0
       AND (MIN(raised_amount) != MAX(raised_amount))
"""
pd.read_sql(query, con = engine).head()
```

8. Создадим поле с категориями:

- Для фондов, которые инвестируют в более 10^9 компаний, назначим категорию large.

- Для фондов, которые инвестируют в 10^6 и более компаний до 10^9, назначим категорию medium.

- Если количество инвестируемых компаний фонда не достигает 10^6, назначим категорию low.

```python
query = """
SELECT
    *,
    CASE
        WHEN raised_amount > 1000000000 THEN 'large'
        WHEN raised_amount BETWEEN 1000000 AND 1000000000 THEN 'medium'
        WHEN raised_amount < 1000000 THEN 'low'
    END as capitalization
FROM fund
"""
pd.read_sql(query, con = engine).sample(5)
```

16. Выгрузим таблицу, в которой будут такие поля:
- название компании-покупателя;
- сумма сделки;
- название компании, которую купили;
- сумма инвестиций, вложенных в купленную компанию;
- доля, которая отображает, во сколько раз сумма покупки превысила сумму вложенных в компанию инвестиций, округлённая до ближайшего целого числа.

Не будем учитывайть те сделки, в которых сумма покупки равна нулю. Если сумма инвестиций в компанию равна нулю, исключим такую компанию из таблицы. Отсортируем таблицу по сумме сделки от большей к меньшей, а затем по названию купленной компании в лексикографическом порядке. Ограничим таблицу первыми десятью записями.

```python
query = """
SELECT c.name AS acquiring_company,
       a.price_amount,
       c_1.name AS acquired_company,
       c_1.funding_total,
       ROUND(a.price_amount/CAST(c_1.funding_total AS float)) AS percent
FROM acquisitions AS a 
LEFT JOIN company AS c ON a.acquiring_company_id = c.id
LEFT JOIN company AS c_1 ON a.acquired_company_id = c_1.id
WHERE a.price_amount>0 
 AND CAST(c_1.funding_total AS float)>0
ORDER BY a.price_amount DESC, acquired_company
LIMIT 10
"""
pd.read_sql(query, con = engine)
```

17. Выгрузим таблицу, в которую войдут названия компаний из категории social, получившие финансирование с 2010 по 2013 год включительно. Проверим, что сумма инвестиций не равна нулю. Выведите также номер месяца, в котором проходил раунд финансирования.

```python
query = """
SELECT
    company.name,
    EXTRACT(MONTH FROM CAST(funding_round.funded_at AS date)) AS month
FROM funding_round
LEFT JOIN company ON company.id=funding_round.company_id
WHERE company.category_code='social'
      AND (EXTRACT(YEAR FROM CAST(funding_round.funded_at AS date)) BETWEEN '2010' AND '2013')
"""
pd.read_sql(query, con = engine).head()
```

18. Отберём данные по месяцам с 2010 по 2013 год, когда проходили инвестиционные раунды. Сгруппируем данные по номеру месяца и получим таблицу, в которой будут поля:
- номер месяца, в котором проходили раунды;
- количество компаний, купленных за этот месяц;
- общая сумма сделок по покупкам в этом месяце.

```python
query = """
WITH month_fund AS
  (SELECT EXTRACT(MONTH
                  FROM fr.funded_at) AS MONTH
   FROM funding_round AS fr
   LEFT JOIN investment AS i ON i.funding_round_id = fr.id
   LEFT JOIN fund AS f ON CAST(i.fund_id AS varchar) = CAST(f.id AS varchar)
   LEFT JOIN company AS c ON CAST(i.fund_id AS varchar) = CAST(c.id AS varchar)
   WHERE EXTRACT(YEAR
                 FROM fr.funded_at) BETWEEN 2010 AND 2013
                 AND c.country_code = 'USA'
                 GROUP BY MONTH),
   month_acquired AS
   (SELECT EXTRACT(MONTH
                  FROM acquired_at) AS MONTH,
          COUNT(acquired_company_id) AS count_of_acquired,
          SUM(price_amount) AS sum_of_acquired
   FROM acquisitions
   WHERE EXTRACT(YEAR
                 FROM acquired_at) BETWEEN 2010 AND 2013
   GROUP BY MONTH)

SELECT month_fund.month,
       month_acquired.count_of_acquired,
       month_acquired.sum_of_acquired
FROM month_fund
JOIN month_acquired ON month_fund.month = month_acquired.month;
"""
pd.read_sql(query, con = engine)
```

19. Составим сводную таблицу и выведем среднюю сумму инвестиций для стран, в которых есть стартапы, зарегистрированные в 2011, 2012 и 2013 годах. Отсортируем таблицу по среднему значению инвестиций за 2011 год от большего к меньшему.

```python
query = """
WITH

    year_2011 AS
                (SELECT 
                     country_code,
                     AVG(CAST(funding_total as float)) AS funding_total_2011
                 FROM company
                 WHERE EXTRACT(YEAR FROM CAST(founded_at AS date)) = 2011
                 GROUP BY country_code),
        year_2012 AS
                (SELECT 
                     country_code,
                     AVG(CAST(funding_total as float)) AS funding_total_2012
                 FROM company
                 WHERE EXTRACT(YEAR FROM CAST(founded_at AS date)) = 2012
                 GROUP BY country_code),
         year_2013 AS
                (SELECT 
                     country_code,
                     AVG(CAST(funding_total as float)) AS funding_total_2013
                 FROM company
                 WHERE EXTRACT(YEAR FROM CAST(founded_at AS date)) = 2013
                 GROUP BY country_code)
                 
                 
SELECT
    year_2011.country_code,
    year_2011.funding_total_2011,
    year_2012.funding_total_2012,
    year_2013.funding_total_2013
FROM year_2011
INNER JOIN year_2012 ON year_2012.country_code=year_2011.country_code
INNER JOIN year_2013 ON year_2013.country_code=year_2012.country_code
ORDER BY year_2011.funding_total_2011 DESC
"""
pd.read_sql(query, con = engine).head()
```
